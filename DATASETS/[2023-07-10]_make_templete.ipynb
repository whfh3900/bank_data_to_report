```python
from itertools import *
from tqdm import tqdm
```

```python
dataset = ['Nic', 'Noun', 'Name', 'Alpha', 'Number']
dataset = ['Nic', 'Noun', 'Name', 'Alpha']


all_list = list()
for i in tqdm(range(1,5)):
    printList = list(product(dataset, repeat = i))
    for j in printList:
        if set(j) - set(['Alpha', 'Noun']):
            all_list.append(j)
            
all_list, len(all_list)
```

```python
f = open('../data/twitter_templates0', 'w', encoding='utf-8-sig')
for i in all_list:
    f.write('\n')
    for j in i:
        f.write(j+' ')
f.close()
```
